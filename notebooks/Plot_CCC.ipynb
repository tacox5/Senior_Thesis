```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import glob
```

```python
files = glob.glob('ccc*')
files.sort()
```

```python
colors = cm.viridis(np.linspace(0.1, 0.9, len(files)))[::-1]
```

```python
ks = np.load('ks.npy')
```

```python
def extract_z(f):
    return float(f.split('_')[-1].rsplit('.', 1)[0])
```

```python
plt.figure(figsize = (10, 6))

for i, f in enumerate(files):
    data = np.load(f)
    plt.scatter(ks, data, label = 'z = {}'.format(extract_z(f)), color = colors[i])
    
plt.xscale('log')
plt.xlim([4e-2, 10])
plt.ylim([-1, 1])
plt.axhline(0, color = 'k', linestyle = ':')
plt.legend(fontsize = 12)
plt.xlabel(r'k [Mpc$^{-1}$]', fontsize = 16)
plt.ylabel(r'CCC$_{\rm 21, Ly\alpha}$(k)', fontsize = 16)
plt.show()
```
